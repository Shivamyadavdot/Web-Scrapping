```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from selenium import webdriver
```

```python
#Opening the browser
driver = webdriver.Chrome()  

name = input("Enter the SYMBOL of stock ")
url = "https://finance.yahoo.com/quote/{}.NS/key-statistics?p={}.NS"

driver.get(url.format(name,name))
driver.implicitly_wait(10)

#retrieveing the source code
html = driver.page_source
#creating the soup
soup = bs(html, 'html.parser')

# Find the div tag with the key statistics
key_stats_div = soup.find('div', {'id': 'Col1-0-KeyStatistics-Proxy'})


#Iterating thorugh the key stats
key_stats_var = key_stats_div.find_all("td")
data = []
for rows in key_stats_var:
    data.append(rows.get_text())
    #print(rows.get_text())

# Find the name, price and change
stock = soup.find(class_ ="D(ib) Fz(18px)").get_text()
price = soup.find(class_="Fw(b) Fz(36px) Mb(-4px) D(ib)").get_text()
change = soup.find(class_ ="Fw(500) Pstart(8px) Fz(24px)").get_text()
driver.quit()

#creating dataframe 
columns = data[::2]
Values  = [data[1::2]]
df = pd.DataFrame(Values,columns=columns)
data = {'Index': [0],
        'Name': stock,
        'Price': price,
        'Change': change
        }
df1 = pd.DataFrame(data)
stock = pd.concat([df1,df],axis =1)
stock
```
